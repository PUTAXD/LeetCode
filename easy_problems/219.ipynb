```python
from typing import List
nums = [1,2,3,1]
k = 1
```

```python
class Solution:
    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        for i in range(len(nums)):
            for j in range(i+1, min((i+1+k),len(nums))):
                if nums[i] == nums[j]:
                    return True
        return False
    
    def containNearbyDuplicateHasmap(self, nums:List[int], k: int) -> bool:
        last_seen = {}
        for i, num in enumerate(nums):
            if num in last_seen and i - last_seen[num] <= k: 
                return True
            last_seen[num] = i 
            print(last_seen)
        return False
```

```python
Solution.containsNearbyDuplicate(Solution(), nums, k)
```
