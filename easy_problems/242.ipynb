```python
from typing import List
s = "anagram"
t = "nagaram"
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        for i in range(len(s)):
            for j in range(len(t)):
                if s[i] == t[j] : 
                    break
            return False

    def isAnagramSort(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        
        s_sorted = sorted(s)
        t_sorted = sorted(t)

        for i in range(len(s)):
            if s_sorted[i] != t_sorted[i] : 
                return False
        
        return True

    def isAnagramHash(sel, s: str, t: str) -> bool : 
        if len(s) != len(t) : 
            return False
        
        count_s, count_t = {},{}

        for i in range(len(s)):
            count_s[s[i]] = 1 + count_s.get(s[i],0)
            count_t[t[i]] = 1 + count_t.get(t[i],0)
        return count_s == count_t
```
