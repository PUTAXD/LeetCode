```python
from typing import List
nums = [1,1,1,3,3,4,3,2,4,2]
```

```python
class Solution:
    def containsDuplicateBruteForce(self, nums: List[int]) -> bool:           
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] == nums[j] : 
                    return True
        return False
    
    def containDuplicateHashSet(self, nums: List[int]) -> bool: 
        s = set()
        for num in nums:
            if num in s:
                return True 
            s.add(num)
        return False
    
    def containDuplicateSorting(self, nums: List[int]) -> bool:
        nums.sort()
        for i in range(len(nums)-1):
            if nums[i] == nums[i+1]:
                return True
        return False
    
    def hasDuplicate(self, nums:List[int]) -> bool:
        return len(set(nums)) < len(nums)
```

```python
Solution.containsDuplicateBruteForce(Solution(), nums)
```

```python
Solution.containDuplicateHashSet(Solution(), nums)
```

```python
Solution.containDuplicateSorting(Solution(), nums=[1,2,3,4,5])
```

```python
Solution.hasDuplicate(Solution(), nums=[1,2,3,4,1])
```
